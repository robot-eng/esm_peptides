```python
import py3Dmol

# Read the PDB file
with open('/content/DMPfold2/dmpfold/example/seq.pdb', 'r') as file:
    pdb_data = file.read()

# Create a viewer
view = py3Dmol.view(width=800, height=600)
view.addModel(pdb_data, 'pdb')
view.setStyle({'cartoon': {'color': 'spectrum'}})
view.zoomTo()
view.show()
```
